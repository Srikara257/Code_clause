```python
import tkinter as tk
import random
import time

# List of sentences for the typing test
sentences = [
    "The quick brown fox jumps over the lazy dog.",
    "Python is a high-level programming language.",
    "Practice makes perfect.",
    "Coding is fun and challenging.",
    "Keep calm and code on."
]

# Function to start the typing test
def start_typing_test():
    global start_time
    start_time = time.time()
    sentence = random.choice(sentences)
    sentence_label.config(text=sentence)
    input_entry.delete(0, tk.END)
    input_entry.focus_set()
    start_button.config(state=tk.DISABLED)

# Function to check the typed text
def check_typing():
    user_input = input_entry.get()
    original_sentence = sentence_label.cget("text")
    if user_input == original_sentence:
        elapsed_time = time.time() - start_time
        words_per_minute = len(original_sentence.split()) / (elapsed_time / 60)
        result_label.config(text=f"Words per minute: {words_per_minute:.2f}")
    else:
        result_label.config(text="Try again.")
    start_button.config(state=tk.NORMAL)

# Create the main window
window = tk.Tk()
window.title("Speed Typing Test")

# Label for displaying the sentence to type
sentence_label = tk.Label(window, text="", font=("Arial", 14))
sentence_label.pack(pady=10)

# Entry field for typing
input_entry = tk.Entry(window, font=("Arial", 12))
input_entry.pack(pady=10)

# Button to start the typing test
start_button = tk.Button(window, text="Start Typing Test", command=start_typing_test, font=("Arial", 12))
start_button.pack()

# Button to check the typed text
check_button = tk.Button(window, text="Check", command=check_typing, font=("Arial", 12))
check_button.pack()

# Label for displaying the result
result_label = tk.Label(window, text="", font=("Arial", 14))
result_label.pack(pady=10)

# Run the main loop
window.mainloop()
```
